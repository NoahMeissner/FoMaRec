## Analysis of Kochbar Dataset

```python
import pandas as pd
import numpy as np
from data_structure.paths import DATA_DIR
import matplotlib.pyplot as plt
from clean_dataset import DataClean
import pandas as pd
import numpy as np
from data_structure.paths import DATA_DIR
import matplotlib.pyplot as plt
```

```python
PATH = DATA_DIR / 'dataset.csv'
PATH_KETO = DATA_DIR / 'Ketodietrecipes.csv'
```

```python
data_clean_obj = DataClean(pd.read_csv(PATH))
df = data_clean_obj.preprocess_data()
df_KETO = pd.read_csv(PATH_KETO)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Prepare grouped data
data_grouped = [
    [df['fat'].dropna(), df_KETO['fat_in_grams'].dropna()],
    [df['carbohydrates'].dropna(), df_KETO['carbohydrates_in_grams'].dropna()],
    [df['protein'].dropna(), df_KETO['protein_in_grams'].dropna()]
]

nutrients = ['Fat', 'Carbohydrates', 'Protein']
positions = np.arange(len(nutrients)) * 3  # spread groups
width = 0.6  # width of each box

plt.figure(figsize=(10, 6))

for i, (normal, keto) in enumerate(data_grouped):
    plt.boxplot(normal, positions=[positions[i] - 0.4], widths=width,
                patch_artist=True, boxprops=dict(facecolor='skyblue'),
                medianprops=dict(color='black'), labels=['Normal'] if i == 0 else [""])
    plt.boxplot(keto, positions=[positions[i] + 0.4], widths=width,
                patch_artist=True, boxprops=dict(facecolor='lightgreen'),
                medianprops=dict(color='black'), labels=['Keto'] if i == 0 else [""])

# X-axis labeling
plt.xticks(positions, nutrients)
plt.ylabel('Amount per Recipe (grams)')
plt.ylim(0, 100)
plt.title('Macronutrient Comparison per Recipe: Normal vs Keto')
plt.legend(["Normal", "Keto"], loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig("keto_vs_normal.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 5))
plt.boxplot(df['kcal'].dropna())
plt.title('Calories Distribution')
plt.ylabel('Calories per Recipe')
plt.ylim(0, 1000)
plt.tight_layout()
plt.show()
```
