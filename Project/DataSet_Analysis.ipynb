## Analysis of Kochbar Dataset

```python
import pandas as pd
import numpy as np
from data_structure.paths import DATA_DIR
import matplotlib.pyplot as plt
from clean_dataset import DataClean
```

```python
PATH = DATA_DIR / 'dataset.csv'
```

```python
data_clean_obj = DataClean(pd.read_csv(PATH))
df = data_clean_obj.preprocess_data()
```

```python
len(df)
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

# Daten vorbereiten
data = [
    df['fat'].dropna(),
    df['carbohydrates'].dropna(),
    df['protein'].dropna()
]

labels = ['Fat (g)', 'Carbohydrates (g)', 'Protein (g)']

# Boxplot erstellen
plt.figure(figsize=(8, 5))
plt.boxplot(data, labels=labels)
plt.title('Macronutrient Distribution')
plt.ylabel('Amount per Recipe (grams)')
plt.ylim(0, 100)  

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 5))
plt.boxplot(df['kcal'].dropna())
plt.title('Calories Distribution')
plt.ylabel('Calories per Recipe')
plt.ylim(0, 1000)
plt.tight_layout()
plt.show()
```
