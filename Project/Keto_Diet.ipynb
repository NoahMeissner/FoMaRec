The keto (ketogenic) diet is a low-carbohydrate, high-fat eating plan that shifts the body into a metabolic state called ketosis, where it burns fat for energy instead of carbohydrates. It typically involves consuming around 70–80% of daily calories from fat, 10–20% from protein, and only 5–10% from carbs.

```python
import pandas as pd
import numpy as np
from data_structure.paths import DATA_DIR
import matplotlib.pyplot as plt
```

```python
PATH = DATA_DIR / 'Ketodietrecipes.csv'
```

```python
df = pd.read_csv(PATH)
```

```python
len(df)
```

### Primary Analysis

```python

import matplotlib.pyplot as plt

# Daten gruppieren nach Kategorie
categories = df['category'].unique()
data = [df[df['category'] == cat]['prep_time_in_minutes'].dropna() for cat in categories]

# Boxplot erstellen
plt.figure(figsize=(10, 6))
plt.boxplot(data, labels=categories)
plt.title('Boxplot Cooking Time by Category')
plt.ylabel('minutes')
plt.ylim(0, 30)  # Skala auf 0–50 Minuten begrenzen
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Median je Kategorie berechnen und ausgeben
for cat in categories:
    median = df[df['category'] == cat]['prep_time_in_minutes'].median()
    print(f"Median cooking time for '{cat}': {median} minutes")
```

```python
import matplotlib.pyplot as plt

category_counts = df['category'].value_counts()

plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar')
plt.title('Recipes per Category')
plt.xlabel('cartegorie')
plt.ylabel('quantity')
plt.xticks()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Daten vorbereiten
data = [
    df['fat_in_grams'].dropna(),
    df['carbohydrates_in_grams'].dropna(),
    df['protein_in_grams'].dropna()
]

labels = ['Fat (g)', 'Carbohydrates (g)', 'Protein (g)']

# Boxplot erstellen
plt.figure(figsize=(8, 5))
plt.boxplot(data, labels=labels)
plt.title('Macronutrient Distribution')
plt.ylabel('Amount per Recipe (grams)')
plt.ylim(0, 100)  # Skala auf 0–50 Minuten begrenzen

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 5))
plt.boxplot(df['calories'].dropna())
plt.title('Calories Distribution')
plt.ylabel('Calories per Recipe')
plt.ylim(0, 1000)
plt.tight_layout()
plt.show()
```
